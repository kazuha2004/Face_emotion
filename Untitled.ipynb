```python
import cv2
from deepface import DeepFace

# Load the Haar cascade for face detection
faceCascade = cv2.CascadeClassifier(r"D:\Git Projects\Emotion detection\haarcascade_frontalface_default.xml")
if faceCascade.empty():
    raise IOError("Error loading Haar cascade XML file.")

# Initialize webcam
cap = cv2.VideoCapture(1)
if not cap.isOpened():
    raise IOError("Cannot open webcam")

# Set higher resolution for better accuracy
cap.set(cv2.CAP_PROP_FRAME_WIDTH, 1280)  # Increase width
cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 720)  # Increase height

try:
    frame_count = 0  # To limit emotion detection to every second frame
    while True:
        ret, frame = cap.read()
        if not ret:
            print("Failed to grab frame")
            break

        # Skip emotion detection on every frame, run it on every second frame
        if frame_count % 2 == 0:
            try:
                # Analyze emotions in the frame
                results = DeepFace.analyze(frame, actions=['emotion'], enforce_detection=False)

                # Check if the result is a list or a single dictionary
                if isinstance(results, list):
                    dominant_emotion = results[0]['dominant_emotion']  # Get the first result
                else:
                    dominant_emotion = results['dominant_emotion']  # Single result case
            except Exception as e:
                dominant_emotion = "Unknown"  # Fallback if analysis fails
                print(f"Emotion detection error: {e}")

        # Convert frame to grayscale for face detection
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        faces = faceCascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=4, minSize=(30, 30))

        # Draw rectangles and display emotion
        for (x, y, w, h) in faces:
            cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)
            cv2.putText(frame, dominant_emotion, (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 0, 255), 2)

        # Display the video feed with rectangles and emotions
        cv2.imshow('Emotion Detection', frame)

        # Exit on pressing Enter (key code 13)
        if cv2.waitKey(1) == 13:  # 13 is the key code for Enter
            print("Exiting...")
            break

        frame_count += 1  # Increment frame counter

except KeyboardInterrupt:
    print("Stopped by user")

# Release resources
cap.release()
cv2.destroyAllWindows()
```
